```python
%matplotlib inline
import os

os.environ['USE_TF'] = '1'
from doctr.io import DocumentFile
from doctr.models import ocr_predictor

predictor = ocr_predictor(pretrained=True)
```

```python
doc = DocumentFile.from_images("op.jpeg")
```

```python
result = predictor(doc)
result.show(doc)
```

```python
json_export = result.export()
print(json_export)
```
